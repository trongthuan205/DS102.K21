## 0 . Import libaray

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

## 1 . Load data

```
! git clone https://github.com/trongthuan205/DS102.K21 DS102.K21
```

```
data_train = pd.read_csv('DS102.K21/House_Train.csv')
data_test = pd.read_csv('DS102.K21/House_Test.csv')
```

# 2 . Split dataset

```
y_train = data_train['MEDV']
y_test = data_test['MEDV']
```

# 3 . Buil model for each variable and Visualization

```
for i in data_test.columns[0:-1]:
    X_train = np.array(data_train[i]).reshape(-1,1)
    lm = LinearRegression()
    lm.fit(X_train, y_train)
    X_test = np.array(data_test[i]).reshape(-1,1)
    y_test_pred = lm.predict(X_test)
    print("R(MEDV, %s) = " % (i), r2_score(y_test, y_test_pred))
    plt.scatter(X_test, y_test, color = "red")
    plt.plot(X_test, y_test_pred, color = 'blue')    
    plt.title("MEDV vs %s" % (i))
    plt.xlabel("%s" % (i))
    plt.ylabel("MEDV")
    plt.show()
```
