# Euclid

```
from numpy.linalg import norm
def distance_euclid(p, q):
    return norm(p - q)
```

```
import numpy as np
```

```
p1 = np.array([0.4,0.53])
p2 = np.array([0.22,0.38])
p3 = np.array([0.353, 0.32])
p4 = np.array([0.26, 0.19])
p5 = np.array([0.08,0.41])
p6 = np.array([0.45,0.30])
```

```
def find_distance_euclid(p, q):
    distances = np.zeros((len(p),len(q)))
    for i in range(len(distances)):
        for j in range (len(distances)):
            distances[i][j] = distance_euclid(p[i],q[j])
    return distances
```

```
p = (p1,p2,p3,p4,p5,p6)
q = (p1,p2,p3,p4,p5,p6)
```

```
print("Khoang cach Euclid: \n",find_distance_euclid(p, q))
```

# Manhatta

```
from scipy.spatial import distance
def distance_manhattan(p, q):
    return distance.cityblock(p, q)
```

# Minkowski

```
def distance_minkowski(p, q, r):
    return distance.minkowski(p, q, r)
```

# Chessboard

```
def distance_chessboard(p, q):
    return distance.chebyshev(p, q)
```

# Hamming

```
def distance_hamming(p, q):
    return distance.hamming(p, q)
```

# Standardized euclidean

```
from scipy.spatial.distance import pdist
def distance_seuclid(p, q):
    return pdist([p, q], "seuclidean", V = None)
```

# Canberra

```
def distance_canberra(p, q):
    return distance.canberra(p, q)
```

# Bray-curtis

```
def distance_braycurtis(p, q):
    return distance.braycurtis(p, q)
```

# Cosin

```
def distance_cosine(p, q):
    return distance.cosine(p, q)
```

# Pearson

```
def distance_pearson(p, q):
    return distance.pearsonr(p, q)
```

# Jaccard

```
def distance_jaccard(p, q):
    return distance.jaccard(p, q)
```
