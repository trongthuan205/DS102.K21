```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from math import sqrt
import joblib
```

```
! git clone https://github.com/trongthuan205/DS102.K21 DS102.K21
```

```
#data 01
train_01 = pd.read_csv('DS102.K21/SalesData_Train01.csv')
test_01 = pd.read_csv('DS102.K21/SalesData_Test01.csv')
X_01_train = np.array(train_01.iloc[:,1].values).reshape(-1,1)
y_01_train = np.array(train_01.iloc[:,2].values)
X_01_test = np.array(test_01.iloc[:,1].values).reshape(-1,1)
y_01_test = np.array(test_01.iloc[:,2].values)

lm_01 = LinearRegression()
lm_01.fit(X_01_train, y_01_train)
y_pred_lm_01 = lm_01.predict(X_01_train)
y_pred_lm_test_01 = lm_01.predict(X_01_test)
joblib.dump(lm_01, 'data_01(linearmodel).sav')

poly_transform_01 = PolynomialFeatures(degree = 4)
X_poly_01 = poly_transform_01.fit_transform(X_01_train)
X_poly_01_test = poly_transform_01.fit_transform(X_01_test)
poly_lm_01 = LinearRegression()
poly_lm_01.fit(X_poly_01, y_01_train)
y_pred_pm_01 = poly_lm_01.predict(X_poly_01)
y_pred_pm_test_01 = poly_lm_01.predict(X_poly_01_test)
joblib.dump(poly_lm_01, 'data_01(polynomial).sav')

R2_lm_01 = r2_score(y_01_test, y_pred_lm_test_01)
RMSE_lm_01 = sqrt(mean_squared_error(y_01_test, y_pred_lm_test_01))
R2_pm_01 = r2_score(y_01_test, y_pred_pm_test_01)
RMSE_pm_01 = sqrt(mean_squared_error(y_01_test, y_pred_pm_test_01))

plt.figure(figsize = (20,5))
plt.subplot(1,3,1)
plt.scatter(X_01_train, y_01_train)
plt.xlabel('Temperature (*C)')
plt.ylabel('Revenue (dollars)')
plt.title('Temperature vs Revenue (Training data 01)')
plt.subplot(1,3,2)
plt.scatter(X_01_train, y_01_train)
plt.plot(X_01_train, y_pred_lm_01, color = 'red')
plt.xlabel('Temperature (*C)')
plt.ylabel('Revenue (dollars)')
plt.title('Temperature vs Revenue (Linear model data 01)')
plt.subplot(1,3,3)
plt.scatter(X_01_train, y_01_train)
plt.plot(X_01_train, y_pred_pm_01, color = 'red')
plt.xlabel('Temperature (*C)')
plt.ylabel('Revenpue (dollars)')
plt.title('Temerature vs Revenue (Polynomial model data 01)')
plt.show()
```
