```
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from math import sqrt
```

```
sum_r2 = 0
sum_RSME = 0
```

```
for i in range(1,6):
    X_train = pd.read_csv('DS102.K21/Bike_Train0'+str(i)+'.csv').iloc[:,3:-2].values
    y_train = pd.read_csv('DS102.K21/Bike_Train0'+str(i)+'.csv').iloc[:,-2].values
    X_test = pd.read_csv('DS102.K21/Bike_Test0'+str(i)+'.csv').iloc[:,3:-2].values
    y_test = pd.read_csv('DS102.K21/Bike_Test0'+str(i)+'.csv').iloc[:,-2].values
    lm = LinearRegression()
    lm = lm.fit(X_train, y_train)
    y_pred_train = lm.predict(X_train)
    y_pred_test = lm.predict(X_test)
    r2 = r2_score(y_test, y_pred_test)
    sum_r2 += r2
    RSME = sqrt(mean_squared_error(y_test, y_pred_test))
    sum_RSME += RSME
    print('R2_%s:' % i, r2)
    print('RMSE_%s:' % i, RSME)
```

```
print('Average of R2: ', sum_r2/5)
print('Average of RMSE: ', sum_RSME/5)
```
