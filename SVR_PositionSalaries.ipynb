#0. Import library

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from math import sqrt
```

#1. Load data

```
train_data = pd.read_csv('DS102.K21/Position_SalariesTrain.csv')
test_data = pd.read_csv('DS102.K21/Position_SalariesTest.csv')
```

```
test_data.head()
```

```
X_train = train_data.iloc[:,1:-1].values
y_train = train_data.iloc[:,-1].values.reshape(-1,1)
X_test = test_data.iloc[:,1:-1].values
y_test = test_data.iloc[:,-1].values.reshape(-1,1)
```

#3. Normalize data

```
SC_X = StandardScaler()
SC_y = StandardScaler()
```

```
X_trans_train = SC_X.fit_transform(X_train)
y_trans_train = SC_y.fit_transform(y_train)
```

#4. Build and Train model

```
svr = SVR(kernel = 'rbf')
svr.fit(X_trans_train, y_trans_train)
```

#5. Redifination

```
def predict(model, X, SC_X, SC_y):
    X_trans = SC_X.transform(X)
    y_trans_pred = model.predict(X_trans)
    y_pred = SC_y.inverse_transform(y_trans_pred)
    return y_pred
```

```
y_pred_train = predict(svr, X_train, SC_X, SC_y)
```

#6. Visualize

```
import matplotlib.pyplot as plt
plt.figure(figsize = (20,5))
plt.subplot(1,3,1)
plt.scatter(X_train, y_train, color = 'red')
plt.xlabel('Position Level')
plt.ylabel('Sallary(dollars/year)')
plt.title('Position vs Salary')
plt.subplot(1,3,2)
X_dumpy_train = X_train.reshape(-1,1)
y_dumpy_pred_train = predict(svr, X_dumpy_train, SC_X, SC_y)
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, y_pred_train, color = 'blue')
plt.xlabel('Position Level')
plt.ylabel('Sallary(dollars/year)')
plt.title('Position vs Salary (Train data)')
plt.subplot(1,3,3)
X_dumpy_test = X_test.reshape(-1,1)
y_dumpy_pred_test = predict(svr, X_dumpy_test, SC_X, SC_y)
plt.scatter(X_dumpy, y_dumpy_pred, color = 'red')
plt.plot(X_test, y_pred_test, color = 'blue')
plt.xlabel('Position Level')
plt.ylabel('Sallary(dollars/year)')
plt.title('Position vs Salary (Test data)')
plt.show()
```

#7. Evaluate

```
print('RMSE  in train: ', sqrt(mean_squared_error(y_train, y_pred_train)))
print('RMSE  in test: ', sqrt(mean_squared_error(y_test, y_pred_test)))
print('MSE  in train: ', mean_squared_error(y_train, y_pred_train))
print('MSE  in test: ', mean_squared_error(y_test, y_pred_test))
print('R2 in train: ', r2_score(y_train, y_pred_train))
print('R2 in test: ', r2_score(y_test, y_pred_test))
```
