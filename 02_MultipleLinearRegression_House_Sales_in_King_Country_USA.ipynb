# 0 . Import library

```
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from math import sqrt
import sklearn.metrics as metrics
from sklearn.metrics import r2_score
```

# 1 . Load data

```
! git clone https://github.com/trongthuan205/DS102.K21 DS102.K21
```

```
data_train = pd.read_csv('DS102.K21/kc_house_Train.csv')
data_test = pd.read_csv('DS102.K21/kc_house_Test.csv')
```

# 2 . Split data

```
X_train = data_train.iloc[:,4:].values
y_train = data_train.iloc[:,3].values
X_test = data_test.iloc[:,4:].values
y_test = data_test.iloc[:,3].values
```

# 3 . Buil model

```
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
y_pred = lin_reg.predict(X_train)
```

# 4 . Comapare

```
def compare(i_example):
    X = X_test[i_example:i_example + 1]
    y = y_test[i_example]
    y_pred = lin_reg.predict(X)
    print(y,y_pred)
```

```
for i in range(len(X_test)):
    compare(i)
```

# 5 . Evaluate

```
y_test_pred = lin_reg.predict(X_test)
```

```
print('RMS=', sqrt(mean_squared_error(y_test, y_test_pred)))
print('MSE= ', mean_squared_error(y_test, y_test_pred))
print('MAE= ', metrics.mean_absolute_error(y_test, y_test_pred))
print('R2= ', r2_score(y_test, y_test_pred))
```
