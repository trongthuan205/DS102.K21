# 0 . Import library

```
import cv2
import glob
import imageio 
import numpy as np
import matplotlib.pyplot as plt
```

# 1 . Load data

```
X_train = []
X_test = []
y_train = []
y_test = []
```

```
for im_path in glob.glob('/content/drive/My Drive/Datasets/COVID/CT_COVID/CT_COVID/*'):
    cv_img = cv2.imread(im_path);     
    gray = cv2.cvtColor(cv_img, cv2.COLOR_BGR2GRAY)   
    img = cv2.resize(gray,(28,28))   
    X_train.append(img)
    y_train.append(1)
```

```
for im_path in glob.glob('/content/drive/My Drive/Datasets/COVID/CT_NonCOVID/CT_NonCOVID/*'):
    cv_img = cv2.imread(im_path);     
    gray = cv2.cvtColor(cv_img, cv2.COLOR_BGR2GRAY)   
    img = cv2.resize(gray,(28,28))   
    X_train.append(img)
    y_train.append(0)
```

```
X_train = np.array(X_train)
```

```
plt.imshow(X_train[0])
```

```
for im_path in glob.glob('/content/drive/My Drive/Datasets/COVID/covidTest/*'):
    cv_img = cv2.imread(im_path);     
    gray = cv2.cvtColor(cv_img, cv2.COLOR_BGR2GRAY)   
    img = cv2.resize(gray,(28,28))   
    X_test.append(img)
    y_test.append(1)
```

```
for im_path in glob.glob('/content/drive/My Drive/Datasets/COVID/NoncovidTest/*'):
    cv_img = cv2.imread(im_path);     
    gray = cv2.cvtColor(cv_img, cv2.COLOR_BGR2GRAY)   
    img = cv2.resize(gray,(28,28))   
    X_test.append(img)
    y_test.append(0)
```
