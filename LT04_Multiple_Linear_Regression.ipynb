# 0 . Import library

```
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
```

# 1 . Load data

```
! git clone https://github.com/trongthuan205/DS102.K21 DS102.K21
```

```
data_train = pd.read_csv('DS102.K21/50_Startups_Train.csv')
data_test = pd.read_csv('DS102.K21/50_Startups_Test.csv')
```

# 2 . Split data

```
X_train = data_train.iloc[:,0:5].values
y_train = data_train.iloc[:,-1].values
```

```
X_test = data_test.iloc[:,0:5].values
y_test = data_test.iloc[:,-1].values
```

# 3 . Preprocessing data

```
ohe_state_train = np.array(pd.get_dummies(X_train[:,-1]))
X_train = np.concatenate((X_train[:,0:3], ohe_state_train), axis = 1)
```

```
X_train[0]
```

```
ohe_state_test = np.array(pd.get_dummies(X_test[:,-1]))
X_test = np.concatenate((X_test[:,0:3], ohe_state_test), axis = 1)
```

```
X_test[0]
```

# 4 . Build model

```
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

# 5 . Evaluate

```
print(lin_reg.score(X_train, y_train))
```

```
print(lin_reg.score(X_test, y_test))
```

# 5 . Compare

```
def compare(i_example):
    X = X_test[i_example:i_example+1,:]
    y = y_test[i_example]
    y_pred = lin_reg.predict(X)[0]
    print(X, y, y_pred )
```

```
for i in range(len(X_test)):
    compare(i)
```
