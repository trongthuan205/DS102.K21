# 0 . Import library

```
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

# 1 . Load data

```
! git clone https://github.com/trongthuan205/DS102.K21 DS102.K21
```

```
data = pd.read_csv('DS102.K21/50_Startups.csv')
```

```
data.head()
```

# 2 . Preprocessing data

```
X = np.array(data.iloc[:,0:4].values)
y = np.array(data.iloc[:,-1].values)
```

```
ohe_state = np.array(pd.get_dummies(X[:,3]))
ohe_state[0]
```

```
X = np.concatenate((X[:,0:3], ohe_state), axis = 1)
X[0]
```

# 4 . Buid model

```
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 0)
```

```
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

# 5 . Comapare 

```
def compare(i_example):
    X = X_test[i_example : i_example + 1]
    y = y_test[i_example]
    y_pred = lin_reg.predict(X)[0]
    print(X, y, y_pred)
```

```
for i in range(len(X_test)):
    compare(i)
```
