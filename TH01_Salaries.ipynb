# 0 . Import library

```
import pandas as pd
import numpy as np
from scipy.spatial.distance import pdist
from numpy.linalg import norm
from scipy.spatial import distance
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# 1 . Load data

```
! git clone https://github.com/trongthuan205/DS102.K21 DS102.K21
```

```
dataset = pd.read_csv('DS102.K21/Salary_Data.csv')
```

# 2 . Split data

```
X = np.array(dataset.iloc[:,0].values).reshape(-1,1)
Y = np.array(dataset.iloc[:,1].values)
```

# 3 . Build model and Visualize

```
for i in np.arange (0.5, 1, 0.1):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = i, random_state = 0)
    # Visualize training data
    plt.figure(figsize = (15,5))
    plt.subplot(1,3,1)
    plt.scatter(X_train, Y_train, color = 'red')
    plt.title("Salary vs Experiment")
    plt.xlabel("Experiment (years)")
    plt.ylabel("Salary (dollars/year)")
    #Visualize training set
    regressor = LinearRegression()
    regressor.fit(X_train, Y_train)
    Y_train_pred = regressor.predict(X_train)
    plt.subplot(1,3,2)
    plt.scatter(X_train, Y_train, color = 'red')
    plt.plot(X_train, Y_train_pred, color = 'blue')
    plt.title("Salary vs Experiment (Training set)")
    plt.xlabel("Experiment (years)")
    plt.ylabel("Salary (dollars/year)")
    #Visualize testing set
    Y_test_pred = regressor.predict(X_test)
    print("Khoang cach Euclid: ", norm(Y_test_pred - Y_test))
    print("Khoang cach Hamming: ", distance.hamming(Y_test_pred, Y_test))
    print("Khoang cach Standarized Euclid: ",  pdist([Y_test_pred, Y_test], "seuclidean", V = None))
    plt.subplot(1,3,3)
    plt.scatter(X_test, Y_test)
    plt.plot(X_test, Y_test, color = 'blue')
    plt.plot(X_test, Y_test_pred, color = 'black')
    plt.title("Salary vs Experiment (Testing set)")
    plt.xlabel("Experiment (years)")
    plt.ylabel("Salary (dollars/year)")
    plt.show()
```
