# 0 . Import libaray

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from math import sqrt
import sklearn.metrics as metrics
from sklearn.metrics import r2_score
```

# 1 . Load data

```
! git clone https://github.com/trongthuan205/DS102.K21 DS102.K21
```

```
data_train = pd.read_csv('DS102.K21/kc_house_Train.csv')
data_test = pd.read_csv('DS102.K21/kc_house_Test.csv')
```

# 2 . Split dataset

```
X_train = data_train.iloc[:,4:]
y_train = data_train.iloc[:,3]
```

```
X_test = data_test.iloc[:,4:]
y_test = data_test.iloc[:,3]
```

# 3 . Visualize train model and Evaluation

```
attribute = ['bedrooms','bathrooms','sqft_living','sqft_lot','floors','sqft_above','sqft_basement']
```

```
for attr in attribute:
    lin_reg = LinearRegression()
    X = np.array(X_train[attr].values).reshape(-1,1)
    lin_reg.fit(X, y_train)
    y_pred = lin_reg.predict(X)
    plt.scatter(X, y_train, color = 'red')
    plt.plot(X,y_pred, color = 'blue')
    plt.xlabel('%s' % attr)
    plt.ylabel('price')
    plt.title('%s vs price(model)' % attr)
    plt.show()
    print('RMS(%s,price) = ' % attr, sqrt(mean_squared_error(y_train, y_pred)))
    print('MSE(%s,price) = ' % attr, mean_squared_error(y_train, y_pred))
    print('MAE(%s,price) = ' % attr, metrics.mean_absolute_error(y_train, y_pred))
    print('R2(%s,price) = ' % attr, r2_score(y_train, y_pred))
```
