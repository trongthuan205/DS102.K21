# 0 . Import library

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
```

# 1 . Load data

```
! git clone https://github.com/trongthuan205/DS102.K21 DS102.K21
```

```
data = pd.read_csv('DS102.K21/Position_Salaries.csv')
```

```
X = data.iloc[:,1:-1].values
y = data.iloc[:,-1].values.reshape(-1,1)
```

# 3 . Build and Traing model

```
model = RandomForestRegressor(n_estimators = 10, random_state = 0)
```

```
model.fit(X, y)
```

# 4 . Visualize

```
plt.figure(figsize = (25,5))
plt.subplot(1,3,1)
plt.scatter(X, y, color = 'red')
plt.title('Postion Level vs Salary')
plt.xlabel('Position Level')
plt.ylabel('Salary (dollars/year)')
plt.subplot(1,3,2)
X_dummy = np.arange(0,10,0.1).reshape(-1,1)
y_dummy_pred = model.predict(X_dummy)
plt.scatter(X, y, color = 'red')
plt.plot(X_dummy, y_dummy_pred, color = 'blue')
plt.title('Position Level vs Salary')
plt.xlabel('Posion Level')
plt.ylabel('Sallary (dollars/year)')
```

# 5 . Compare

```
def compare(i_example):
    x = X[i_example : i_example + 1]
    Y = y[i_example]
    y_pred = model.predict(x)
    print(x, Y, y_pred)
```

```
for i in range(len(X)):
    compare(i)
```
