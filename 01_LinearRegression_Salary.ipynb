# 0 . Import libaray

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

# 1 . Load data

```
! git clone https://github.com/trongthuan205/DS102.K21 DS102.K21
```

```
dataset = pd.read_csv('DS102.K21/Salary_Data.csv')
```

# 2 . Split dataset

```
X = np.array(dataset.iloc[:,0].values).reshape(-1,1)
Y = np.array(dataset.iloc[:,1].values)
```

```
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.8, random_state = 0)
```

# 3 . Train model

```
lin_reg = LinearRegression()
lin_reg.fit(X_train, Y_train)
```

# 4 . Visualize

```
# Visualize training data
plt.figure(figsize = (20,5))
plt.subplot(1,3,1)
plt.scatter(X_train, Y_train, color = 'red')
plt.title("Salary vs Experiment")
plt.xlabel("Experiment (years)")
plt.ylabel("Salary (dollars/year)")
#Visualize training set
Y_train_pred = lin_reg.predict(X_train)
plt.subplot(1,3,2)
plt.scatter(X_train, Y_train, color = 'red')
plt.plot(X_train, Y_train_pred, color = 'blue')
plt.title("Salary vs Experiment (Training set)")
plt.xlabel("Experiment (years)")
plt.ylabel("Salary (dollars/year)")
#Visualize testing set
X_test_dummy = sorted(X_test)
Y_test_dummy = sorted(Y_test)
Y_test_pred = lin_reg.predict(X_test_dummy)
plt.subplot(1,3,3)
plt.scatter(X_test_dummy, Y_test_dummy, color = 'red')
plt.plot(X_test_dummy, Y_test_pred, color = 'blue')
plt.title("Salary vs Experiment (Testing set)")
plt.xlabel("Experiment (years)")
plt.ylabel("Salary (dollars/year)")
plt.show()
```

# 5 . Compare

```
def compare(i_example):
    X = X_test[i_example:i_example + 1]
    y = Y_test[i_example]
    y_pred = lin_reg.predict(X)
    print(X, y, y_pred)
```

```
for i in range(len(X_test)):
    compare(i)
```
